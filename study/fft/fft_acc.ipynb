```python
import numpy as np # numpyをインポート
import matplotlib.pyplot as plt # matplotlibをインポート
import pandas as pd
```

```python
acc = pd.read_csv("acc.csv",encoding = 'utf-8')
print(acc)
```

```python
### 量子化(離散化) ###
f_s = 50 # サンプリングレート f_s[Hz] (任意)
t_fin = acc["time"].iloc[-1] # 収録終了時刻 [s] (任意)
dt = 1/f_s # サンプリング周期 dt[s]
N = int(f_s * t_fin) # サンプル数 [個]
```

```python
### y(t) のグラフ ###
plt.plot(acc["time"], acc["x"]) # y-t グラフのプロット
plt.xlim(0, 20) # 横軸に関する描画範囲指定
plt.show() # グラフの表示
```

```python
### FFT: tの関数をfの関数にする ###
y_fft = np.fft.fft(acc["x"]) # 離散フーリエ変換
freq = np.fft.fftfreq(N, d=dt) # 周波数を割り当てる（※後述）
Amp = abs(y_fft/(N/2)) # 音の大きさ（振幅の大きさ）
```

```python
### 音波のスペクトル ###
plt.plot(freq[1:int(N/2)], Amp[1:int(N/2)]) # A-f グラフのプロット
# plt.xscale("log") # 横軸を対数軸にセット
# plt.xlim([1,10]) #x軸の範囲を1~5に指定
plt.show()
```

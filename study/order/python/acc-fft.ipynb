```python
#import
import matplotlib.pyplot as plt
from matplotlib import dates as mdates

import json
import datetime

#基本的なコード
#フォントサイズの変更
plt.rcParams["font.size"] = 28
plt.rcParams['font.family'] = "IPAexGothic"
```

```python
# csvの読み込み
# ヘッダなしの場合csv = pd.read_csv("ファイル名(.csv)", header=None, names=['user_id', 'name'])で可能
fname = "../data/"

with open(fname+"data2.json", 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)

# JSONデータを辞書として読み込んだので、必要な情報にアクセスできます
order_nums = data.get("orderNums")
print(order_nums)
```

```python
# アイテムごとの売上数を格納する辞書を初期化
sales_data = {}

first = 0

for order_num in order_nums:
    if order_num is not None:
        first = order_num.get("timestamp")
        print(first)
        break

for order_num in order_nums:
    if order_num is not None:
        order_list = order_num.get("orderList")
        timestamp = order_num.get("timestamp")

        # Unixエポックから時間を取得し、X軸の値として使用
        hours = (timestamp - first) / 3600
        if(hours > 7500):
            continue

        if order_list is not None:
            for _, value in order_list.items():
                item = value.get("item")
                qty = value.get("qty")
                if item in sales_data:
                    sales_data[item] += qty
                else:
                    sales_data[item] = qty
```

```python
# グラフ描画
items = list(sales_data.keys())
quantities = list(sales_data.values())

plt.figure(figsize=(10, 12))
plt.bar(items, quantities)
plt.xlabel('アイテム')
plt.ylabel('売上数量')
plt.title('1日目のアイテムごとの売上数量')
plt.xticks(rotation=90, ha='center') 

#画像を保存するときに下を多めにする
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.4, top=0.9)

# グラフを保存
plt.savefig(f'1日目のアイテムごとの売上数量.png')

plt.show()
```

```python
# アイテムごとの売上数を格納する辞書を初期化
sales_time_data = {}

for order_num in order_nums:
    if order_num is not None:
        order_list = order_num.get("orderList")
        unix_time = order_num.get("timestamp") / 1000
        timestamp = datetime.datetime.fromtimestamp(unix_time)

        print(timestamp)

        if order_list is not None:
            for _, value in order_list.items():
                item = value.get("item")
                qty = value.get("qty")
                if item not in sales_time_data:
                    sales_time_data[item] = {"timestamps": [], "quantities": []}
                
                sales_time_data[item]["timestamps"].append(timestamp)
                sales_time_data[item]["quantities"].append(qty)
```

```python
# 商品ごとにグラフを表示
for item, data in sales_time_data.items():
    timestamps = data["timestamps"]
    quantities = data["quantities"]

    print(item)
    print(timestamps)
    print(quantities)

    # 2023-10-07 00:00:00 からの　2023-10-07 23:59:59までのグラフを描画
    # この時間をX軸の値として使用
    start_time = datetime.datetime(2023, 10, 7, 0, 0, 0)
    end_time = datetime.datetime(2023, 10, 7, 23, 59, 59)

    # quantityの累積和を計算
    cumsum_quantities = []
    cumsum = 0
    for timestamp, qty in zip(timestamps, quantities):
        if start_time <= timestamp <= end_time:
            cumsum += qty
            cumsum_quantities.append(cumsum)

    timestamps = timestamps[:len(cumsum_quantities)]

    print(cumsum_quantities)

    # グラフを描画
    plt.figure(figsize=(30, 20))
    plt.plot(timestamps, cumsum_quantities)
    plt.xlabel('時間 (Unixエポック)')
    plt.ylabel('売上数量')
    plt.title(f'1日目{item} の売上数量')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

    # グラフを保存
    plt.savefig(f'1日目{item} の売上数量.png')

    plt.show()

```

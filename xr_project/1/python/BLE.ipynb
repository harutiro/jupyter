```python
#import
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt


#基本的なコード
#フォントサイズの変更
plt.rcParams["font.size"] = 28
plt.rcParams['font.family'] = "IPAexGothic"
```

```python
# csvの読み込み
# ヘッダなしの場合csv = pd.read_csv("ファイル名(.csv)", header=None, names=['user_id', 'name'])で可能
fname = "../data/1/"

ble_df = pd.read_csv( # type: ignore
    fname + "BLE.csv",
    encoding='utf-8',
    engine='python',
    dtype={'time':'int', 'id':'str', 'rssi':'int'}
) 
```

```python
# idの値を取得
id_list = list(ble_df["id"].unique())


# idの値ごとにrssiを取得
for i in id_list:
    item : pd.DataFrame = ble_df[ble_df["id"]==i]
    
    # 正規分布の確率密度関数を求める
    try:
        #  平均と標準偏差から確率を求める
        mean = item["rssi"].mean()
        std = item["rssi"].std()
        print(f"mean: {mean}, std: {std}")

        # グラフの描画
        x = range(-100, 0)
        y = (1/(std * (2 * 3.1415) ** 0.5)) * (2.718 ** -((x-mean) ** 2 / (2 * std ** 2)))

        print(f"sum: {sum(y)}")
        print(y)

        plt.plot(x, y, label="normal distribution")
        plt.legend()

        # 正規分布から確率を求める
        print(f"rssi <= -80: {sum(y[0:20])}")
        print(f"rssi <= -90: {sum(y[0:10])}")
        print(f"rssi <= -100: {sum(y[0:0])}")

    except:
        print("error")
        continue

    plt.title(i)
    plt.xlabel("rssi")
    plt.ylabel("density")
    # グラフの表示
    plt.show()
    
```

```python
# idの値を取得
id_list = list(ble_df["id"].unique())


# idの値ごとにrssiを取得
for i in id_list:
    item : pd.DataFrame = ble_df[ble_df["id"]==i]
    
    # 平均と標準偏差を取得
    mean = item["rssi"].mean()
    std = item["rssi"].std()

    fig, ax = plt.subplots(figsize=(15, 6)) 
    # ヒストグラムの描画
    ax.hist(item["rssi"], bins=20, alpha=0.5, label=i)
    ax.set_title('rssiのヒストグラム count = ' + str(len(item["rssi"])))
    ax.set_xlabel('rssi')
    ax.set_ylabel('count')
    ax.legend()
    fig.show()
    
    print(f"id:{i}, mean:{mean}, std:{std}")
```

```python
# idの値を取得
id_list = list(ble_df["id"].unique())

fig, ax = plt.subplots(figsize=(15, 6)) 

# idの値ごとにrssiを取得
for i in id_list:
    item : pd.DataFrame = ble_df[ble_df["id"]==i]
    
    # 平均と標準偏差を取得
    mean = item["rssi"].mean()
    std = item["rssi"].std()

    # グラフにプロット
    ax.scatter(i, mean, c='blue')
    ax.errorbar(i, mean, yerr=std, fmt='o', ecolor='blue', capsize=4)

    print(f"id:{i}, mean:{mean}, std:{std}")

# グラフの表示
plt.xticks(rotation=90)
plt.show()
    
```

```python
# ble_dfのidに対してそれぞれのrssiを配列に格納
# idの値を取得
id_list = list(ble_df["id"].unique())

# idの値ごとにrssiを取得
rssi_dict = {}
for i in id_list:
    item : pd.DataFrame = ble_df[ble_df["id"]==i]
    rssi_dict[item["id"].values[0]] = item["rssi"].values.tolist()

print(rssi_dict)

# グラフの描画
fig, ax = plt.subplots(figsize=(15, 6))  # Set the width to 12 and height to 6
ax.boxplot(x=rssi_dict.values(), labels=rssi_dict.keys())
# x軸のラベルを縦書きにする
plt.xticks(rotation=90)
ax.set_xlabel("id")
ax.set_ylabel("rssi")
plt.show()
```

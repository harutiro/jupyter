# グラフの読み込み

```python
from Graphs import MakeGraph ,GraphAxis , GraphViewOptions

folder_name = "テント生活1"

make_graph = MakeGraph(folder_name=folder_name)
```

# 加速度

```python
acc_df = make_graph.acc_df
acc_norm_df = make_graph.norm(acc_df)

make_graph.plot(
    plt_lists=[
        make_graph.acc_df,
        acc_norm_df
    ],
    option=[GraphViewOptions.none],
    axis=GraphAxis.acc,
    filter_num=50
)
```

# 角速度

```python
gyro_df = make_graph.angular_df
gyro_norm_df = make_graph.norm(gyro_df)

make_graph.plot(
    plt_lists=[
        gyro_df,
        gyro_norm_df
    ],
    option=[GraphViewOptions.none],
    axis=GraphAxis.acc,
    filter_num=50
)
```

# BLE

```python
# 100以上の値を除去
ble_char_df = make_graph.ble_char_df
ble_char_df = ble_char_df[ble_char_df["rssi"] < 100]
ble_tent_df = make_graph.ble_tent_df
ble_tent_df = ble_tent_df[ble_tent_df["rssi"] < 100]

make_graph.ble_plot(
    plt_lists=[
        ble_char_df,
        ble_tent_df
    ],
    axis=GraphAxis.rssi,
    filter_num=50,
    label_name=["char","tent"]
)

```

# HeartRate

```python
heart_rate_df = make_graph.heart_rate_df

make_graph.plot(
    plt_lists=[
        heart_rate_df
    ],
    option=[GraphViewOptions.none],
    axis=GraphAxis.bpm,
    filter_num=50
)
```

# 平均などを出力する

```python
info_df = make_graph.calc(calc_list=acc_df,filter_num=50)

make_graph.output_csv(info_df,file_name=folder_name)
```
